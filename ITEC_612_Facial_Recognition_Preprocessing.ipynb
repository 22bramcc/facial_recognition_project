```python
import os
from pathlib import Path
import cv2
import matplotlib.pyplot as plt


def load_dataset(input_folder):
    """
    Load all image file paths from the dataset folder.

    Args:
        input_folder (str): Path to the folder containing images.

    Returns:
        list: A list of full image file paths.
    """
    image_extensions = {".jpg", ".jpeg", ".png", ".bmp", ".tif", ".tiff"}
    image_paths = []

    for root, _, files in os.walk(input_folder):
        for file in files:
            if Path(file).suffix.lower() in image_extensions:
                image_paths.append(os.path.join(root, file))

    return image_paths


def preprocess_image(image_path, target_size=(128, 128), grayscale=True):
    """
    Read and preprocess a single image.

    Preprocessing steps:
    - Load image
    - Resize image
    - Convert to grayscale (optional)
    - Normalize pixel values to range [0, 1]
    """
    image = cv2.imread(image_path)

    if image is None:
        print(f"Warning: Could not load image: {image_path}")
        return None, None

    image = cv2.resize(image, target_size)

    if grayscale:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    image = image / 255.0
    original_filename = os.path.basename(image_path)
    return image, original_filename


def save_processed_image(image, output_folder, filename):
    """
    Save a processed image to the output folder.
    """
    os.makedirs(output_folder, exist_ok=True)
    save_image = (image * 255).astype("uint8")
    output_path = os.path.join(output_folder, filename)
    cv2.imwrite(output_path, save_image)


def visualize_image(image, title="Processed Image"):
    """
    Display an image using Matplotlib.
    """
    plt.figure(figsize=(4, 4))

    if len(image.shape) == 2:
        plt.imshow(image, cmap="gray")
    else:
        plt.imshow(cv2.cvtColor((image * 255).astype("uint8"), cv2.COLOR_BGR2RGB))

    plt.title(title)
    plt.axis("off")
    plt.show()


def process_dataset(input_folder, output_folder, target_size=(128, 128), grayscale=True, preview=False):
    """
    Process all images in the dataset folder and save the outputs.
    """
    image_paths = load_dataset(input_folder)

    if not image_paths:
        print("No images found in the dataset folder.")
        return

    print(f"Found {len(image_paths)} image(s). Starting preprocessing...")

    for image_path in image_paths:
        processed_image, filename = preprocess_image(
            image_path,
            target_size=target_size,
            grayscale=grayscale
        )

        if processed_image is not None:
            save_processed_image(processed_image, output_folder, filename)

            if preview:
                visualize_image(processed_image, title=filename)

    print("Preprocessing complete. Processed images saved successfully.")
```

```python
input_folder = "dataset"
output_folder = "processed_data"

process_dataset(
    input_folder=input_folder,
    output_folder=output_folder,
    target_size=(128, 128),
    grayscale=True,
    preview=False
)
```
